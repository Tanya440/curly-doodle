```python
import pickle
import json
class customer:
    cus_list=[]
    
    def __init__(self):
        self.id=0
        self.name=0
        self.mob_no=0
        self.age=0
    def addcustomer(self):
        customer.cus_list.append(self)
    def searchcustomer(self):
        for e in customer.cus_list:
            if e.id==self.id:
               return e 
        raise ValueError("ID not found!")
        
    def updatecustomer(self):
        for e in customer.cus_list:
            if e.id==self.id:
                e.name=self.name
                e.mob_no=self.mob_no
                e .age=self.age
            raise ValueError("ID not found!")
    def deletecustomer(self):
        for e in customer.cus_list:
            if e.id ==self.id:
                customer.cus_list.remove(e)
                return
        raise ValueError("Customer not found!")
    @staticmethod
    def savetopickle():
        f=open(r"C:\anaconda2021\progrmaming python\cmspickle.text","wb")
        pickle.dump(customer.cus_list,f) 
        f.close()
    @staticmethod
    def loadFrompickle():
        f=open(r"C:\anaconda2021\progrmaming python\cmspickle.text","rb")
        customer.cus_list=pickle.load(f)
        f.close
    @staticmethod
    def criteriaID(ob):
        return ob.id
    @staticmethod
    def mysort():
        customer.cus_list.sort(key=customer.criteriaID)
        
    @staticmethod
    def convertTodict(ob):
        return ob.__dict__
    @staticmethod
    def saveToJSON():
        f=open(r"C:\anaconda2021\progrmaming python\cmspickle.text","w")
        json.dump(customer.cus_list,f,default=customer.convertTodict)
        f.close()
    @staticmethod
    def convertToobj(d):
        cus=customer()
        cus.id=d['id']
        cus.name=d['name']
        cus.mob_no=d['mob_no']
        cus.age=d['age']
        return cus
    @staticmethod
    def loadFromJSON():
        f=open(r"C:\anaconda2021\progrmaming python\cmspickle.text","r")
        customer.cus_list=json.load(f,object_hook=customer.convertToobj)
        f.close()
from tkinter import *
from tkinter import messagebox
        
def btnAdd_click():
    cus=customer()
    cus.id=varID.get()
    cus.name=varName.get()
    cus.mob_no=varMob_no.get()
    cus.age=varAge.get()
    cus.addcustomer()
    messagebox.showinfo("CMS","Customer Added Successfully")
    varID.set("")
    varName.set("")
    varMob_no.set("")
    varAge.set("")
def btnSearch_click():
    cus=customer()
    cus.id=varID.get()
    cus=cus.searchcustomer()
    varName.set(cus.name)
    varMob_no.set(cus.mob_no)
    varAge.set(cus.age)
    
    
def  btnDelete_click():
    ch=messagebox.askyesno("CMS","Are you sur you want to delete this customer? ")
    if ch==True:
        cus=customer()
        cus.id=varID.get()
        cus.deletecustomer()
        messagebox.showinfo("CMS","Customer Deleted Successfully")
def btnModify_click():
    cus=customer()
    cus.id=varID.get()
    cus.name=varName.get()
    cus.mob_no=varMob_no.get()
    cus.age=varAge()
    cus.updatecustomer()
    messagebox.showinfo("CMS","Customer Updated Successfully")
def btnAll_click():
    root1=tk()
    root1.tittle("All Customers")
    lblID1=label(root1,text="CUSTOMER ID",bg="orange",width=10,font=20)
    lblID1.grid(row=0,column=0)
    lblName1=label(root1,text="CUSTOMER NAME",bg="orange",width=10,font=20)
    lblName1.grid(row=0,column=1)
    lblMob1=label(root1,text="CUSTOMER MOBILE NUMBER",bg="orange",width=10,font=20)
    lblMob1.grid(row=0,column=2)
    lblAge1=label(root1,text="CUSTOMER AGE",bg="orange",width=10,font=20)
    lblAge1.grid(row=0,column=3)
    i=1
    for cus in customer.cus_list:
        lblID2=label(root1,text=f"{cus.id}",bg="yellow",width=10,font=20)
        lblID2.grid(row=i,column=0)
        lblName2=label(root1,text=f"{cus.name}",bg="yellow",width=10,font=20)
        lblName2.grid(row=i,column=1)
        lblMob2=label(root1,text=f"{cus.mob_no}",bg="yellow",width=10,font=20)
        lblMob2.grid(row=i,column=2)
        lblAge2=label(root1,text="{cus.age}",bg="yellow",width=10,font=20)
        lblAge2.grid(row=i,column=3)
        i+=1
        

def btnExit_click():
    root.destroy()

    
root=Tk()
root.geometry("600x600")
root.title("CUSTOMER MANAGEMENT SYSTEM")
  
lblID = Label(root, text = "Enter Customer ID:", font = 20)
lblID.grid(row = 0, column = 0)
varID = StringVar()
entrID = Entry(root, textvariable = varID, font = 20)
entrID.grid(row = 0, column = 1)

lblName = Label(root, text = "Enter Customer Name:", font = 20)
lblName.grid(row = 1, column = 0)
varName = StringVar()
entrName = Entry(root, textvariable = varName, font = 20)
entrName.grid(row = 1, column = 1)

lblAge = Label(root, text = "Enter Customer Mobile no:", font = 20)
lblAge.grid(row = 2, column = 0)
varAge = StringVar()
entrAge = Entry(root, textvariable = varAge, font = 20)
entrAge.grid(row = 2, column = 1)

lblMob_no = Label(root, text = "Enter Customer Age:", font = 20)
lblMob_no.grid(row = 3, column = 0)
varMob_no = StringVar()
entrMob_no = Entry(root, textvariable = varMob_no, font = 20)
entrMob_no.grid(row = 3, column = 1)

btnAdd = Button(root, text = "Add Customer", font = 20, width = 15, command = btnAdd_click)
btnAdd.grid(row = 4, column = 0)

btnSearch = Button(root, text = "Search Customer", font = 20, width = 15, command = btnSearch_click)
btnSearch.grid(row = 4, column = 1)

btnDelete = Button(root, text = "Delete Customer", font = 20, width = 15, command = btnDelete_click)
btnDelete.grid(row = 4, column = 2)

btnModify = Button(root, text = "Modify Customer", font = 20, width = 15, command = btnModify_click)
btnModify.grid(row = 5, column = 0)

btnAll = Button(root, text = "View all Cusustomer", font = 20, width = 15, command = btnAll_click)
btnAll.grid(row = 5, column = 1)

btnExit = Button(root, text = "Exit", font = 20, width = 15, command = btnExit_click)
btnExit.grid(row = 5, column = 2)


root.mainloop()
     

```
